```python
import pandas as pd
# MUA Index
MUA_index_clean = pd.read_csv('Resources/MUA_index_clean.csv')
# Incidence
age_inc = pd.read_csv('Resources/age_inc.csv')
race_inc = pd.read_csv('Resources/race_inc.csv')
sex_inc = pd.read_csv('Resources/sex_inc.csv')
# Mortality 
age_mort = pd.read_csv('Resources/age_mort.csv')
race_mort = pd.read_csv('Resources/race_mort.csv')
sex_mort = pd.read_csv('Resources/sex_mort.csv')
# Population
population_state = pd.read_csv('Resources/population_state.csv')
# Census Population 2015 to 2019
census_population_2015_2019 = pd.read_csv('Resources/census_population_2015_2019.csv', encoding='latin-1')
# State Aabbreviation
state_abbreviation = pd.read_csv('Resources/state_abbreviation.csv')
```

```python
print(state_abbreviation.shape)
state_abbreviation.drop(['Abbrev'], 1, inplace=True)
state_abbreviation = state_abbreviation.rename(columns={"Code":"Abbrv"})
state_abbreviation.head()
```

```python
print(census_population_2015_2019.shape)
census_population_2015_2019.drop(['Unnamed: 0'], 1, inplace=True)
census_population_2015_2019 = census_population_2015_2019.rename(columns={"STNAME":"State", "CTYNAME":"County"})
census_population_2015_2019.head()
```

```python
census_population_2015_2019['County'] = census_population_2015_2019['County'].str.strip()
```

```python
statesonly = ["Alabama", "Alaska", "Arizona", "Arkansas", "California",
                       "Colorado", "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana",
                       "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota",
                       "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey", "New Mexico",
                       "New York", "North Carolina", "North Dakota", "Ohio", "Oklahoma" ,"Oregon", "Pennsylvania", "Rhode Island",
                       "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington",
                       "West Virginia", "Wisconsin", "Wyoming"]
census_population_2015_2019 = census_population_2015_2019[census_population_2015_2019.County.isin(statesonly) == False]
census_population_2015_2019.head()
```

```python
print(census_population_2015_2019.columns.tolist())
```

```python
state_county = pd.merge(census_population_2015_2019, state_abbreviation)
state_county['County_State'] = state_county['County'].astype(str) +", "+ state_county['Abbrv']
state_county.drop(['POPESTIMATE2015', 'POPESTIMATE2016', 'POPESTIMATE2017', 'POPESTIMATE2018', 'POPESTIMATE2019', 'Abbrv'], 1, inplace=True)
print(state_county.shape)
state_county.head()
```

```python
print(census_population_2015_2019.shape)
census_population_2015_2019 = pd.merge(state_county, census_population_2015_2019)
census_population_2015_2019.head()
```

```python
census_population_2015_2019.drop(['County'], 1, inplace=True)
census_population_2015_2019.head()
```

```python
print(MUA_index_clean.shape)
MUA_index_clean.drop(['Unnamed: 0'], 1, inplace=True)
MUA_index_clean.head()
```

```python
print(age_inc.shape)
age_inc.drop(['Unnamed: 0'], 1, inplace=True)
age_inc['County'] = age_inc['County'].str.replace('\d+', '')
age_inc['County'] = age_inc['County'].str.strip()
age_inc = pd.merge(age_inc, state_county)
age_inc.head()
```

```python
age_inc.drop(['County'], 1, inplace=True)
age_inc.head()
```

```python
print(age_inc.shape)
```

```python
print(race_inc.shape)
race_inc.drop(['County'], 1, inplace=True)
race_inc.head()
```

```python
print(sex_inc.shape)
sex_inc['County'] = sex_inc['County'].str.replace('\d+', '')
sex_inc['County'] = sex_inc['County'].str.strip()
sex_inc = pd.merge(sex_inc, state_county)
sex_inc.head()
```

```python
print(sex_inc.shape)
sex_inc.drop(['County'], 1, inplace=True)
sex_inc.head()
```

```python
print(age_mort.shape)
age_mort['County'] = age_mort['County'].str.replace('\d+', '')
age_mort['County'] = age_mort['County'].str.strip()
age_mort = pd.merge(age_mort, state_county)
age_mort.head()
```

```python
print(age_mort.shape)
age_mort.drop(['County'], 1, inplace=True)
age_mort.head()
```
