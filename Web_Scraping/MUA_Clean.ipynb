```python
import pandas as pd
df = pd.read_csv('MUA_index.csv')
df = df.drop('Expand', axis='columns')
df
```

```python
df.columns
```

```python
#Rename Columns
df = df.rename(columns={"Discipline ":"Discipline",
 "MUA/P ID ":"MUA/P_ID",
 "Service Area Name ":"Service_Area_Name",
 "Designation Type ":"Designation_Type",
 "Primary State Name ":"Primary_State_Name",
 "County ":"County",
 "Index of Medical Underservice Score ":"Index_of_Medical_Underservice_Score",
 "Status ":"Status",
 "Rural Status ":"Rural_Status",
 "Designation Date ":"Designation_Date",
 "Update Date ":"Update_Date"})
df.head()
```

```python
#In the original file there are two rows between each line item with expanded data we dont need
df = df.dropna()
df
```

```python
# Drop Service Area Name because the county name is in "County" column.
    #"Service Area" column has other unwanted text or misspelled text.
# Drop MUA/P_ID because once the combined counties are split there will be duplicate ID
# Drop "Status" it creates duplicate counties
# Drop "Rural_Status" it creates duplicate counties
# Drop date columns because they are not needed
df.drop(['Discipline', 'MUA/P_ID','Service_Area_Name', 'Designation_Type', 'Status', 'Rural_Status', 'Designation_Date', 'Update_Date'], 1, inplace=True)
df.head()
```

```python
#Row count BEFORE splitting combined counties
print(df.shape)
```

```python
df = (df.set_index(df.columns.drop('County',1).tolist())
     .County.str.split('|', expand=True)
     .stack()
     .reset_index()
     .rename(columns={0:'County'})
     .loc[:, df.columns]
    )
df.head()
#One of the duplicate County rows was on line 29, df.head(33) to show the new rows
```

```python
#Row count AFTER splitting combined counties. 
    #In the Alabama data set there were 2 rows with multiple columns. The row count is now increased by 2
print(df.shape)
```

```python
df.dtypes
```

```python
# Change "Index_of_Medical_Underservice_Score" data type fro object to int
df['Index_of_Medical_Underservice_Score'] = df['Index_of_Medical_Underservice_Score'].astype(float).astype(int, errors='ignore')
```

```python
df.dtypes
```

```python
#Strip blank spaces before and after strings
df["County"] = df["County"].str.strip()
df["Primary_State_Name"] = df["Primary_State_Name"].str.strip()
```

```python
# Get Average "Index_of_Medical_Underservice_Score" per "County"
mua_score_df = df.groupby('County', as_index=False)['Index_of_Medical_Underservice_Score'].mean()
mua_score_df.head()
```

```python
# Drop "Index_of_Medical_Underservice_Score" from original df
df.drop(['Index_of_Medical_Underservice_Score'], 1, inplace=True)
df.head()
```

```python
# Drop duplicate counties
df = df.drop_duplicates()
df = df.drop([1], axis=0)
df = df.reset_index(drop=True)
df.head()
```

```python
print(df.shape)
```

```python
df2_merged = pd.merge(df, mua_score_df)
df2_merged
```

```python
print(df2_merged.shape)
```

```python
df2_merged.to_csv('MUA_index_clean.csv')
```
